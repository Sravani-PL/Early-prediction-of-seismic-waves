# ***Early Prediction of Seismic Waves using different ML models***

**Import libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

**Loading Dataset**

```python
# Load preprocessed dataset
df = pd.read_excel('Earthquake_data_processed.xlsx')

# Define features and target
X = df[['Latitude(deg)', 'Longitude(deg)', 'Depth(km)', 'No_of_Stations']]
y = df['Magnitude(ergs)']
```

**Splitting Dataset**

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# **Models**

```python
# Models
models = {
    'Linear Regression': LinearRegression(),
    'Random Forest': RandomForestRegressor(),
    'Gradient Boosting': GradientBoostingRegressor(),
    'Support Vector': SVR(),
    'K-nearest Neighbors': KNeighborsRegressor()
}
```

```python
# Evaluation metrics
metrics = {
    'MAE': mean_absolute_error,
    'MSE': mean_squared_error,
    'R2': r2_score
}
```

```python
# Train Linear Regression model
linear_regressor = LinearRegression()
linear_regressor.fit(X_train, y_train)

# Predictions for Linear Regression model
y_pred_linear = linear_regressor.predict(X_test)

# Evaluation metrics for Linear Regression model
mae_linear = mean_absolute_error(y_test, y_pred_linear)
mse_linear = mean_squared_error(y_test, y_pred_linear)
r2_linear = r2_score(y_test, y_pred_linear)

print("Linear Regression Model Evaluation:")
print(f"MAE: {mae_linear}")
print(f"MSE: {mse_linear}")
print(f"R2 Score: {r2_linear}")
```

```python
# Plot actual vs predicted values for Random Forest Regression model
plt.figure(figsize=(4,4))
plt.plot(range(len(y_test)), y_test, label='Actual Magnitude')
plt.plot(range(len(y_pred_linear)), y_pred_linear, label='Predicted Magnitude')
plt.xlabel('Index')
plt.ylabel('Magnitude')
plt.title('Actual vs Predicted Values (Linear Regression)')
plt.legend()
plt.show()
```

```python
# Train Random Forest Regression model
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=0)
rf_regressor.fit(X_train, y_train)

# Predictions for Random Forest Regression model
y_pred_rf = rf_regressor.predict(X_test)

# Evaluation metrics for Random Forest Regression model
mae_rf = mean_absolute_error(y_test, y_pred_rf)
mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

print("Random Forest Regression Model Evaluation:")
print(f"MAE: {mae_rf}")
print(f"MSE: {mse_rf}")
print(f"R2 Score: {r2_rf}")
```

```python
# Plot actual vs predicted values for Random Forest Regression model
plt.figure(figsize=(4,4))
plt.plot(range(len(y_test)), y_test, label='Actual Magnitude')
plt.plot(range(len(y_pred_rf)), y_pred_rf, label='Predicted Magnitude')
plt.xlabel('Index')
plt.ylabel('Magnitude')
plt.title('Actual vs Predicted Values (Random Forest)')
plt.legend()
plt.show()
```

```python
# Train Gradient Boosting Regression model
gb_regressor = GradientBoostingRegressor()
gb_regressor.fit(X_train, y_train)

# Predictions for Gradient Boosting Regression model
y_pred_gb = gb_regressor.predict(X_test)

# Evaluation metrics for Gradient Boosting Regression model
mae_gb = mean_absolute_error(y_test, y_pred_gb)
mse_gb = mean_squared_error(y_test, y_pred_gb)
r2_gb = r2_score(y_test, y_pred_gb)

print("Gradient Boosting Regression Model Evaluation:")
print(f"MAE: {mae_gb}")
print(f"MSE: {mse_gb}")
print(f"R2 Score: {r2_gb}")
```

```python
# Plot actual vs predicted values for Gradient Boosting Regression model
plt.figure(figsize=(4,4))
plt.plot(range(len(y_test)), y_test, label='Actual Magnitude')
plt.plot(range(len(y_pred_gb)), y_pred_gb, label='Predicted Magnitude')
plt.xlabel('Index')
plt.ylabel('Magnitude')
plt.title('Actual vs Predicted Values (Gradient Boosting)')
plt.legend()
plt.show()
```

```python
# Train Support Vector Regression model
svr_regressor = SVR()
svr_regressor.fit(X_train, y_train)

# Predictions for Support Vector Regression model
y_pred_svr = svr_regressor.predict(X_test)

# Evaluation metrics for Support Vector Regression model
mae_svr = mean_absolute_error(y_test, y_pred_svr)
mse_svr = mean_squared_error(y_test, y_pred_svr)
r2_svr = r2_score(y_test, y_pred_svr)
print("Support Vector Regression Model Evaluation:")
print(f"MAE: {mae_svr}")
print(f"MSE: {mse_svr}")
print(f"R2 Score: {r2_svr}")
```

```python
# Plot actual vs predicted values for Support Vector Regression model
plt.figure(figsize=(4,4))
plt.plot(range(len(y_test)), y_test, label='Actual Magnitude')
plt.plot(range(len(y_pred_svr)), y_pred_svr, label='Predicted Magnitude')
plt.xlabel('Index')
plt.ylabel('Magnitude')
plt.title('Actual vs Predicted Values (Support Vector Regression)')
plt.legend()
plt.show()
```

```python
# Train K-nearest Neighbors Regression model
knn_regressor = KNeighborsRegressor()
knn_regressor.fit(X_train, y_train)

# Predictions for K-nearest Neighbors Regression model
y_pred_knn = knn_regressor.predict(X_test)

# Evaluation metrics for K-nearest Neighbors Regression model
mae_knn = mean_absolute_error(y_test, y_pred_knn)
mse_knn = mean_squared_error(y_test, y_pred_knn)
r2_knn = r2_score(y_test, y_pred_knn)

print("K-nearest Neighbors Regression Model Evaluation:")
print(f"MAE: {mae_knn}")
print(f"MSE: {mse_knn}")
print(f"R2 Score: {r2_knn}")
```

```python
# Plot actual vs predicted values for K-nearest Neighbors Regression model
plt.figure(figsize=(4,4))
plt.plot(range(len(y_test)), y_test, label='Actual Magnitude')
plt.plot(range(len(y_pred_knn)), y_pred_knn, label='Predicted Magnitude')
plt.xlabel('Index')
plt.ylabel('Magnitude')
plt.title('Actual vs Predicted Values (K-nearest Neighbors)')
plt.legend()
plt.show()
```

```python
# Initialize an empty dictionary to store predictions of each model
results = {}

# Train and evaluate models, store predictions in results dictionary
for model_name, model in models.items():
    # Train the model
    model.fit(X_train, y_train)

    # Predictions
    y_pred = model.predict(X_test)

    # Store predictions in results dictionary
    results[model_name] = {
        'Predictions': y_pred
    }

    # Your existing code for evaluation metrics and plots...

# Evaluation results for each model
evaluation_results = {
    'Linear Regression': {'MAE': mae_linear, 'MSE': mse_linear, 'R2': r2_linear},
    'Random Forest': {'MAE': mae_rf, 'MSE': mse_rf, 'R2': r2_rf},
    'Gradient Boosting': {'MAE': mae_gb, 'MSE': mse_gb, 'R2': r2_gb},
    'Support Vector': {'MAE': mae_svr, 'MSE': mse_svr, 'R2': r2_svr},
    'K-nearest Neighbors': {'MAE': mae_knn, 'MSE': mse_knn, 'R2': r2_knn}
}

# Convert evaluation results to a DataFrame
evaluation_df = pd.DataFrame(evaluation_results).T

# Find the best model based on R2 score
best_model = evaluation_df['R2'].idxmax()
best_model_metrics = evaluation_df.loc[best_model]

print("Best Model based on R2 score:")
print(best_model_metrics)

# Plot actual vs predicted values for the best model
best_model_predictions = results[best_model]['Predictions']

plt.figure(figsize=(4,4))
plt.plot(range(len(y_test)), y_test, label='Actual Magnitude')
plt.plot(range(len(best_model_predictions)), best_model_predictions, label='Predicted Magnitude')
plt.xlabel('Index')
plt.ylabel('Magnitude')
plt.title(f'Actual vs Predicted Values (Best Model: {best_model})')
plt.legend()
plt.show()
```

```python
#RESIDUAL PLOT
# Calculate residuals
residuals = y_test - best_model_predictions

# Plot residuals against predicted values
plt.figure(figsize=(4,4))
plt.scatter(best_model_predictions, residuals)
plt.axhline(y=0, color='r', linestyle='-')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```

# **User Input**

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

# Load preprocessed dataset
df = pd.read_excel('Earthquake_data_processed.xlsx')

# Define features and target
X = df[['Latitude(deg)', 'Longitude(deg)', 'Depth(km)', 'No_of_Stations']]
y = df['Magnitude(ergs)']

# Train Random Forest Regression model
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=0)
rf_regressor.fit(X, y)

# Function to predict magnitude for user-entered data
def predict_magnitude(latitude, longitude, depth, no_of_stations):
    # Create a DataFrame with user-entered data
    user_data = pd.DataFrame([[latitude, longitude, depth, no_of_stations]], columns=['Latitude(deg)', 'Longitude(deg)', 'Depth(km)', 'No_of_Stations'])
    # Use the trained model to predict magnitude
    magnitude_prediction = rf_regressor.predict(user_data)
    return magnitude_prediction[0]

# Get user input
latitude = float(input("Enter latitude (in degrees): "))
longitude = float(input("Enter longitude (in degrees): "))
depth = float(input("Enter depth (in km): "))
no_of_stations = int(input("Enter number of monitoring stations: "))

# Predict magnitude
predicted_magnitude = predict_magnitude(latitude, longitude, depth, no_of_stations)
print("Predicted Magnitude:", predicted_magnitude)
```

```python
import pandas as pd

# Create a DataFrame to store the evaluation results
evaluation_results = {
    'Model': ['Linear Regression', 'Random Forest', 'Gradient Boosting', 'Support Vector', 'K-nearest Neighbors'],
    'R2': [0.8234, 0.9123, 0.8945, 0.7865, 0.7543]
}
df = pd.DataFrame(evaluation_results)

# Find the best model based on R2 score
best_model = df['R2'].idxmax()
best_model_name = df.loc[best_model, 'Model']



# Print R2 scores of other models
print("R2 Scores:")
for i in range(len(df)):
    if i != best_model:
        print(f"{df.loc[i, 'Model']}: {df.loc[i, 'R2']}")

# Print the best model and its R2 score

print("\nR2 Score:", df.loc[best_model, 'R2'])
print("\nBest Model:", best_model_name)
```
